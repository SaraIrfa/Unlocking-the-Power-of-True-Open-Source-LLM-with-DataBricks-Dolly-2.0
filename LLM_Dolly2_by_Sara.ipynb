```python
import torch
from transformers import pipeline
```

```python
generate_text = pipeline(model="databricks/dolly-v2-2-8b", torch_dtype=torch.bfloat16, trust_remote_code=True, device_map='auto' )
```

```python
generate_text('Provide instructions for 1 year old baby diet plan')
```

```python
generate_text('How to handle cough and muccus for 3 years old boy')
```

```python
generate_text('My 3 years old boy is showing a lot tantrums, how to handle him?')
```

```python
generate_text('My AIM is to help community by my data science techniques, ML Projects, How can i help others by my work?')
```

```python
generate_text('Give me some school lunch ideas for my three years old boy, he is a picky eater')
```

```python
generate_text('But i have to cook at home')
```
